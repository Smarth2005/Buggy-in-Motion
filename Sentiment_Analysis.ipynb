```python
from transformers import pipeline
```

```python
sentiment_analyzer = pipeline("sentiment-analysis") # default model : distilbert-base-uncased-finetuned-sst-2-english
                                                    # Binary Sentiment Classification (Positive / Negative)
```

```python
texts=[
    "I love cricket!",
    "I hate when Virat Kohli misses a century.",
    "This restaurant has the best food I've ever tasted.",
    "I absolutely love this product, it exceeded all my expectations!",
    "I am disappointed, the product broke within a week.",
    "I regret buying this, it doesn't work as advertised",
    "I have a meeting scheduled for 3 PM tomorrow.",
    "It rained yesterday, and today it's sunny.",
    "I like the taste of the coffee, though it's a bit expensive.",
    "The phone's camera is amazing, but the battery drains too quickly.",
    "The college fest was so much fun, everyone enjoyed the cultural night.",
    "The Wi-Fi in the hostel never works when I need it the most.",
    "Sessionals are so tiring, they drain all the energy before the main exams.",
    "Her smile feels like the sunshine on the darkest day."
]

results = sentiment_analyzer(texts)

for text,result in zip(texts,results):
  print(f"Text: {text}")
  print(f"Predicted Sentiment: {result['label']}, score={result['score']:.2f}")
  print()
```

```python
# Three-Class Sentiment (Positive / Neutral / Negative)
model2 = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment-latest")

results = model2(texts)

for text,result in zip(texts,results):
  print(f"Text: {text}")
  print(f"Predicted Sentiment: {result['label']}, score={result['score']:.2f}")
  print()
```

```python
# Aspect-Based Sentiment Analysis (ABSA)
model3 = pipeline("sentiment-analysis", model="yangheng/deberta-v3-base-absa-v1.1")

results = model3(texts)

for text,result in zip(texts,results):
  print(f"Text: {text}")
  print(f"Predicted Sentiment: {result['label']}, score={result['score']:.2f}")
  print()
```
